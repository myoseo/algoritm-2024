### 1)문제 정의


```python
문제이름: 정렬되지 않은 리스트를 크기순으로 정렬하기
문제설명: 정렬되지 않은 리스트를 비교하여 크기순으로 리스트를 정렬한다.
문제예시 : 
list : [5, 3, 8, 4, 9, 1, 6, 2, 7]
다음 리스트를 정렬하면 아래의 리스트가 나온다.
list : [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

### 2)알고리즘 설명

```python
명칭: 리스트 정렬하기
입력: 정렬되지 않은 리스트
출력: 정렬된 리스트
처리순서:
1. i를 리스트의 가장 마지막 인덱스까지 반복하게 하여 리스트의 i번째 수와 그 다음 숫자들을 비교 하여 가장 작은 수를 비교한다.
2. 가장 작은 수를 least에 넣어준다
3. 가장 작은 수가 들어간 least변수를 리스트의 i번째인덱스에 저장한다.
```

### 4)알고리즘 개요(입력 변수, 출력, 알고리즘 개요)

```python
1) 자료 구조 정의
- 정렬되지 않은 숫자 리스트 :  A(int []) - 정수 배열
- 임의로 작은값을 저장할 리스트 인덱스 : least - 정수
- 리스트의 길이 : n - 정수

2) 함수 정의
- 기능: 리스트를 크기순으로 정렬
- 입력값: 정렬되지 않은 리스트 A(int [])
- 반환값: 정렬된 리스트 A(int [])
- 함수이름: selection_sort(A)
```

### 5)알고리즘 코드

```python
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i
        for j in range(i+1, n):
            if(A[j]<A[least]:
                least = j
        A[i], A[least] = A[least],A[i]
        printfstep(A, i+1)
```

### 8) 복잡도 분석

```python
알고리즘은 n번 반복함.
각 반복 단계에서는 리스트의 나머지 부분 (n-i 요소)에 대해 가장 작은 요소를 찾아냄.
각 반복 단계에서 n-i 요소에 대해 i번 비교를 수행.
첫 번째 반복에서는 n-1번의 비교를 수행하고, 두 번째 반복에서는 n-2번의 비교를 수행. 
이것을 같이 n-1번 반복.
따라서 전체 비교 횟수는 n-1 + n-2 + ... + 1 = (n-1) * n / 2가 .
결국 이 알고리즘의 복잡도는 O(n^2)가 됨.
```
