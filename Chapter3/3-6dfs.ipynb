### 4)알고리즘 개요(입력 변수, 출력, 알고리즘 개요)

1) 자료 구조 정의
- 한 정점을 지났는지 구분해주는 변수: start 
- 차진합 연산의 값을 저장해줄 변수 : nbr
- 방문한 노드를 넣어줄 리스트 : graph[]

2) 함수 정의
- 기능: 깊이우선 탐색을 사용하여 그래프를 탐색해 방문한 정점 순서를 출력해줌
- 함수이름: def dfs(graph, start, visited)
- 입력값: 그래프를 구현한 딕셔너리 graph, 임의 시작지점 A, 거친 정점을 넣어주는 리스트 visited
- 반환값: 제공된 그래프를 깊이 우선 탐색으로 계산한 지나친 정점 순서

### 5)알고리즘 코드 

```python
def dfs(graph, start, visited):
    if start not in visited:
        visited.add(start)
        print(start, end=' ')
        nbr = graph[start] - visited
        for v in nbr:
            dfs(graph, v, visited)
```

### 6)테스트 코드

```python
def dfs(graph, start, visited):
    if start not in visited:
        visited.add(start)
        print(start, end=' ')
        nbr = graph[start] - visited
        for v in nbr:
            dfs(graph, v, visited)

mygraph = {"A" : {"B", "C"},
            "B" : {"A", "D"},
            "C" : {"A", "D", "E"},
            "D" : {"B", "C", "F"},
            "E" : {"C", "G", "H"},
            "F" : {"D"},
            "G" : {"E", "H"},
            "H" : {"E", "G"}
            }
print('DFS :',end='')
dfs(mygraph, "A", set())
print()
```

### 8)복잡도 분석

방문된 노드 수는 최대 N개일때
DFS 함수에서는 모든 노드를 한 번씩 방문하기 때문에 총 n번과 e번의 간선을 지남
e는 n보다 값이 낫기때문에 이 알고리즘의 복잡도는 O(n)이다.
