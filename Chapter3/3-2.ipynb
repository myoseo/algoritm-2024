# 1. 문제정의
- 문제: 순차 탐색 구현  
- 설명: n개의 항목이 들어있는 리스트에서 "탐색키"를 가진 항목을 찾아라. 찾는 항목이 리스트 내 존재할 경우 index를 반환하고, 존재하지 않을 경우 -1을 반환해라. 단, 리스트의 항목들은 정렬되어있지 않다.  
- 예시: [1, 3, 2, 4] 리스트에 '3'이라는 키는 존재한다. 따라서 index[1]이 반환된다.  

# 2. 알고리즘
이번 챕터에선 억지 기법을 통해 알고리즘을 설계하고 구현한다.  
입력 리스트의 첫 번째 항목부터 마지막 항목까지 탐색키와 비교하는 방법을 사용해 해당 문제를 구현해보려고 한다.  
for 구문을 통해 key값과 리스트[i]의 값을 반복적으로 비교해 값을 리턴한다.

# 4. 코드 개요
함수: sequential_search  
입력 변수: a  
출력 변수: lista  
### 함수(sequential_search)
- 기능: 입력받은 리스트의 len값만큼 반복해 해당 index와 key의 값을 비교해 값을 리턴한다.
1. sequential_search 함수를 통해 리스트의 첫 번째 항목을 key와 비교한다.
2. 마지막 항목까지 key항목과 리스트의 해당 index의 값이 동일한지 비교한다.
3. 만약 key의 값과 리스트A의 index의 값이 동일하면 해당 index를 return한다.
4. 리스트에 key값과 동일한 값이 존재하지 않을 경우 -1을 return한다.

### 입력 변수(a)
- 기능: 사용자로부터 값을 입력받아 저장한다. 해당 변수는 함수에 활용된다.

전체 코드
1. 변수 a에 사용자로부터 찾고 싶은 key값을 입력받아 저장함.
2. 함수 sequential_search를 통해 정의된 lista의 리스트와 변수 a를 key값으로 입력해 값을 리턴받는다.
3. 만약 리턴값이 -1일 경우, 해당 리스트에 key값이 존재하지 않음을 print한다.
4. 리턴값이 다르게 나올 경우, 리스트에 해당 key값이 존재하므로 사용자에게 해당 key값의 index를 print한다.

# 5. 알고리즘 코드

```python
# AL3-2 순차 탐색
def sequential_search(A, key) :
    for i in range(len(A)) :
        if A[i] == key :
            return i
    return -1

lista = [1,3,5,7,9,11,13,15]
a = int(input("찾을 값: "))
if sequential_search(lista, a) == -1 :
    print("찾는 값이 존재하지 않음.")
else :
    print(f"리스트에 {a}이 존재합니다. (인덱스: {sequential_search(lista, a)})")
```

# 6. 테스트 코드
random 모듈을 활용해 리스트를 정의한다. 중복되는 수는 다시 입력한다.

```python
# AL3-2 순차 탐색
import random as r

def sequential_search(A, key) :
    for i in range(len(A)) :
        if A[i] == key :
            return i
    return -1

lista = []

for i in range(10) :
    a = r.randint(1,20)
    while a in lista :
        a = r.randint(1,20)
    lista.append(a)
    
a = int(input("찾을 값: "))
print(f"리스트: {lista}")
if sequential_search(lista, a) == -1 :
    print("찾는 값이 존재하지 않음.")
else :
    print(f"리스트에 {a}이 존재합니다. (인덱스: {sequential_search(lista, a)})")
```

# 8. 복잡도 분석
- 최선의 경우: 만약 리스트의 첫 항목에 해당 key가 바로 존재할 경우, 한 번의 시도로 탐색이 종료된다. 따라서 시간 복잡도 함수는 T(n) = 1이며 O(1)이다.
- 최악의 경우: 만약 찾고자 하는 값이 리스트에 존재하지 않거나 맨 끝에 존재할 경우, 리스트의 길이만큼 해당 함수를 반복해야한다. 따라서 T(n) = n이며 O(n)이다.
- 평균의 경우: 찾고자 하는 값이 리스트의 중간정도에 위치할 경우, T(n) = (n + 1)/2이며 O(n)이다.
