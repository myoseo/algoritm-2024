# 1. 문제정의
- 문제: 그래프 탐색(bfs) 구현
- 설명: queue 모듈을 사용해 너비 우선 탐색을 구현해라.너비 우선 탐색은 시작 정점으로부터 가까운 정점을 먼저 방문하고 멀리 떨어져 있는 정점을 나중에 방문하는 순회 방법이다. 
- 예시: A에서 시작해 B, C를 방문 후 B, C를 큐에 넣는다. 이후 B에서 인접한 변수중 방문하지 않은 D를 방문 후 큐에 D를 넣는다. 이하 반복

# 2. 알고리즘
Queue 구조를 활용해 너비 우선 탐색방식을 구현한다.  
각 정점들을 방문할 때 마다 큐에 인접한 정점들을 삽입하고, 방문할 인접 정점이 없는경우 큐의 앞에서 정점을 꺼내 인접한 정점을 차례대로 방문한다.

# 4. 코드 개요
변수 정의  
- 방문 정점을 저장하는 리스트 visited
- 큐 구조를 활용하기 위한 객체 que  

함수 정의  
- 함수 이름: bfs(graph, start)
- 기능: 너비 우선 탐색기법을 구현한다. 각 정점에 방문해 인접한 정점들을 큐에 저장한다.

# 5. 알고리즘 코드

```python
import queue
def bfs(graph, start) :
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty() :
        v = que.get()
        print(v, end='')
        nbr = sorted(graph[v] - visited)
        for u in nbr:
            visited.add(u)
            que.put(u)
```

# 6. 테스트 코드

```python
import queue
def bfs(graph, start) :
    visited = { start }
    que = queue.Queue()
    que.put(start)
    while not que.empty() :
        v = que.get()
        print(v, end=' ')
        nbr = sorted(graph[v] - visited)
        for u in nbr:
            visited.add(u)
            que.put(u)
            
mygraph = { "A": {"B","C"}, 
           "B": {"A","D"}, 
           "C": {"A","D","E"}, 
           "D": {"B","C","F"}, 
           "E": {"C","G","H"}, 
           "F": {"D"}, 
           "G": {"E","H"}, 
           "H": {"E","G"} }

print('BFS : ', end='')
bfs(mygraph, "A")
print()
```

# 8. 복잡도 분석
- 너비 우선 탐색은 거리가 d인 정점들을 모두 방문한 다음, 거리가 d+1인 정점들을 방문하게 된다.
