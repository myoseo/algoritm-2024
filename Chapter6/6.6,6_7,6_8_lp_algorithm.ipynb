# 1. 문제정의
- 선형 조사법으로 삽입 연산, 탐색 연산, 삭제 연산을 구현해라.

# 2. 알고리즘
- 명칭: 선형 조사법의 삽입, 탐색, 삭제 알고리즘
- 입력: 
  - 정수 key
- 처리 순서: 
  - 삽입 연산  
    1. 입력된 key값을 M으로 나머지 연산을 수행한다.
    2. id번째 테이블이 비어있지 않으면 탐색을 계속 수행한다.
    3. id번째 테이블이 비어있으면 해당 항목에 key를 저장한다.
  - 탐색 연산
    1. 입력된 key값의 버킷이 비었으면 None을 리턴한다.
    2. 만약 버킷의 항목이 존재하고, 입력된 key값과 다른 값이 존재할 경우 다음 버킷을 검사한다.
  - 삭제 연산
    1. 탐색한 버킷에 해당 key값이 존재할 경우, 항목을 삭제ㅏㄴ다.

# 4. 코드 개요
1. 변수
- None으로 초기화된 해시테이블 M
2. 함수  
- hashFn(key) :
  - 입력된 key값을 M으로 나머지 연산을 수행한다.
    - 입력: 정수 key
- lp_insert(key) :
  - 입력된 key값이 테이블에 존재하지 않을 경우 비어있는 버킷에 키를 저장한다.
- lp_search(key) :
  - 입력된 key값의 버킷이 비었을 경우 None을 리턴한다. 만약 해당 버킷이 비어있지 않으며 key값과 항목의 값이 다를 경우 다음 버킷을 검사한다.
- lp_delete(key) :
  - 입력된 key값에 해당하는 버킷의 항목값과 key값이 동일할 경우 해당 항목을 삭제한다.(-1을 return함)

# 5. 알고리즘 코드


```python
M = 13
table = [None]*M

def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None and table[id] != -1) :
        id = (id + 1 + M) % M
        count -= 1
    if count > 0 :
        table[id] = key
    return

def lp_search(key) :
    id = hashFn(key)
    count = M
    while count > 0 :
        if table[id] == None :
            return None
        if table[id] != -1 and table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
        return None
    
def lp_delete(key) :
    id = hashFn(key)
    count = M
    while count >0 :
        if table[id] == None : return
        if table[id] != -1 and table[id] == key :
            table[id] = -1
            return 
        id = (id + 1 + M) % M
        count -= 1
```

# 6. 테스트 코드

```python
M = 13
table = [None]*M

def hashFn(key) :
    return key % M

def lp_insert(key) :
    id = hashFn(key)
    count = M
    while count>0 and (table[id] != None and table[id] != -1) :
        id = (id + 1 + M) % M
        count -= 1
    if count > 0 :
        table[id] = key
    return

def lp_search(key) :
    id = hashFn(key)
    count = M
    while count > 0 :
        if table[id] == None :
            return None
        if table[id] != -1 and table[id] == key :
            return table[id]
        id = (id + 1 + M) % M
        count -= 1
    return None
    
def lp_delete(key) :
    id = hashFn(key)
    count = M
    while count >0 :
        if table[id] == None : return
        if table[id] != -1 and table[id] == key :
            table[id] = -1
            return 
        id = (id + 1 + M) % M
        count -= 1


print("   최초: ", table)
lp_insert(45); print("45 삽입: ", table)
lp_insert(27); print("27 삽입: ", table)
lp_insert(88); print("88 삽입: ", table)
lp_insert(9); print(" 9 삽입: ", table)
lp_insert(71); print("71 삽입: ", table)
lp_insert(60); print("60 삽입: ", table)
lp_insert(46); print("46 삽입: ", table)
lp_insert(38); print("38 삽입: ", table)
lp_insert(24); print("24 삽입: ", table)
lp_delete(60); print("60 삭제: ", table)
print(f"46 탐색:  {lp_search(46)}")
```
