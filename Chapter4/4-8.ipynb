### 1)문제 정의

입력된 x의 n승을 구하시오.

### 2) 알고리즘 설명

명칭: 거듭제곱(축소 정복 기법)
입력: 밑(x), 지수(n)
출력: 밑이 x이고 지수 n인 수의 값
처리순서:
1. 지수인 n이 0인지 확인한다.(0이면 1을 반환)
2. 지수인 n이 짝수인지 확인한다.(짝수면 알고리즘에 x*X, n//2를 입력)
3. 지수인 n이 홀수인지 확인한다.(홀수면 알고리즘에 x*x, (n-1)//2를 입력)
4. 2,3을 1이 반환될 때 까지 반복한다.

### 4) 코드 개요(입력 변수, 출력, 함수 설명)

1) 자료 구조 정의
- 입력된 값을 저장할 변수 :  x, n - 정수


2) 함수 정의
- 기능: 입력된 밑이 x이고 지수가n의 값을 계산하여줌.
- 함수이름: power(x, n):
- 입력값: 밑의 값 x, 지수 값 n
- 반환값: 밑이 x이고 지수가n의 값

### 5) 알고리즘 코드

```python
def power(x, n) :
   if n == 0 :
        return 1
   elif (n%2) == 0 :
        return power(x*x, n//2)
   else :
        return x*power(x*x, (n-1)//2)
```

### 6) 테스트 코드


```python
def power(x, n) :
   if n == 0 :
        return 1
   elif (n%2) == 0 :
        return power(x*x, n//2)
   else :
        return x*power(x*x, (n-1)//2)
x, n = map(int, input("밑과 지수를 입력하시오 >> ").split())
print("밑:", x, "지수", n, power(x, n))
```
