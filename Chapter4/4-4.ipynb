### 1)문제 정의

주어진 그래프에 존재하는 각 정점들의 선행 순서를 위배하지 않으면서 모든 정점들을 거친 순서대로 나열해라.

### 2) 알고리즘 설명 

명칭: 위상 정렬(축소 정복 기법)
입력: 주어진 방향 그래프를 표현한 딕셔너리
출력: 모든 정점들을 거친 순서
처리순서:
1. 방향 그래프에서 진입차수가 0인 정점 중에서 하나를 입의로 선택한다.
2. 선택된 정점을 삭제한다.(정점이 삭제되므로 이 정점에서 진출하는 모든 간선들도 삭제한다.)
3. 간선이 삭제되므로 삭제되는 간선돠 연결된 이웃 정점들의 진입차수를 갱신한다.
4. 위의 과정을 모든 정점이 삭제 될때까지 반복한다.
5. 정점이 삭제되는 순서인 위상 순서를 반환한다.

### 4) 코드 개요(입력 변수, 출력, 함수 설명)

1) 자료 구조 정의
- 방향그래프를 저장할 딕셔너리 :  graph - 딕셔너리
- 거친 정점 저장을 위한 딕셔너리 : indeg - 딕셔너리



2) 함수 정의
- 기능: 축소 정복 기법을 이용하여 주어진 그래프에 존재하는 각 정점들의 선행 순서를 위배하지 않으면서 모든 정점들을 거친 순서대로 나열해줌
- 함수이름: topological_sort(graph)
- 입력값: 주어진 방향 그래프를 표현한 딕셔너리 graph
- 축소 정복 기법을 이용하여 모든 정점들을 거친 순서(위상 순서)

### 5)알고리즘 코드

```python
def topological_sort(graph) :
    inDeg = {}
    for v in graph:
        inDeg[v] = 0
    for v in graph:
        for u in graph[v]:
            inDeg[u] += 1
    
    vlist = []
    for v in graph:
        if inDeg[v] == 0:
            vlist.append(v)
    
    while vlist:
        v = vlist.pop()
        print(v, end=' ')

        for u in graph[v]:
            inDeg[u] -= 1
            if inDeg[u] == 0:

                vlist.append(u)
```
