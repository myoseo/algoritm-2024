# 1) 문제 정의
이진트리의 전위, 중위, 후위 순회 알고리즘을 정의하라.

# 2) 알고리즘 설명
이진트리의 전위 순회는   
[루트노드]  -  [왼쪽트리]  -  [오른쪽트리]   
순으로   
   
이진트리의 중위 순회는   
[왼쪽트리]  -  [루트노드]  -  [오른쪽트리]     
순으로    
   
이진트리의 후위 순회는    
[왼쪽트리]  -  [오른쪽트리]  -  [루트노드]    
순으로   
트리를 방문하는 알고리즘이다.

# 4) 코드 설명 (설계)
#### 1. 자료구조 정의   
클래스 TNode:   
이진트리를 표현하기위해 클래스를 사용한다.   
초기화를 위한 함수인 생성자 메소드 __init__ 을 선언하여   
노드의 데이터, 왼쪽 자식 트리 링크, 오른쪽 자식 트리 링크를 선언한다.    
그리고 클래스가 인스턴스화 될때 사용자가 초기화한 값으로 생성된다.

#### 2. 함수 정의
def preorder(n):   
매개변수: 부모노드 n   
함수내용: 매 노드에서 왼쪽을 우선순위로 간후 오른쪽으로간다.   
def inorder(n):    
매개변수: 부모노드 n   
함수내용: 단말노드를 만날때까지 왼쪽 노드로 쭉 내려간뒤   
         왼쪽 -> 부모 -> 오른쪽 를 반복하며 탐색한다.   
def postorder(n):    
매개변수: 부모노드 n   
함수내용: 단말노드를 만날때까지 왼쪽 노드로 쭉 내려간뒤   
         왼쪽 -> 오른쪽 -> 부모 를 반복하며 탐색한다.   

# 5) 코드

```python
class TNode:                               # 이진트리 노드 클래스
    def __init__(self, data, left, right): # 인스턴스가 생성될 때 초기화 작업을 수행하는 메소드 / 노드의 데이터와 왼쪽, 오른쪽 자식트리를 초기화한다.
        self.data = data
        self.left = left
        self.right = right

d = TNode('D', None, None)    # --------------- 
e = TNode('E', None,None)     #
b = TNode('B', d, e)          #  이진트리 초기화
f = TNode('F', None, None)    #
c = TNode('C', f, None)       # 
root = TNode('A', b, c)       # ----------------

def preorder(n):
    if n is not None:
        print(n.data, end = ' ')
        preorder(n.left)
        preorder(n.right)

def inorder(n):
    if n is not None:        
        inorder(n.left)
        print(n.data, end = ' ')
        inorder(n.right)

def postorder(n):
    if n is not None:        
        postorder(n.left)
        postorder(n.right)
        print(n.data, end = ' ')

print('전위 순회: ')
preorder(root)
print('\n')
print('중위 순회: ')
inorder(root)
print('\n')
print('후위 순회: ')
postorder(root)
```

# 6) 테스트 코드 (다른 트리로)

```python
class TNode:                               # 이진트리 노드 클래스
    def __init__(self, data, left, right): # 인스턴스가 생성될 때 초기화 작업을 수행하는 메소드 / 노드의 데이터와 왼쪽, 오른쪽 자식트리를 초기화한다.
        self.data = data
        self.left = left
        self.right = right

e = TNode('E', None,None)     #  --------------- 
b = TNode('B', d, e)          #
j = TNode('J', None, None)    #  
f = TNode('F', j, None)       #
g = TNode('G', None, None)    #  이진트리 초기화                      
h = TNode('H', None, None)    #
i = TNode('I', None, None)    #
d = TNode('D', h, i)          #
c = TNode('C', f, g)          #
root = TNode('A', b, c)       # ----------------

def preorder(n):
    if n is not None:
        print(n.data, end = ' ')
        preorder(n.left)
        preorder(n.right)

def inorder(n):
    if n is not None:        
        inorder(n.left)
        print(n.data, end = ' ')
        inorder(n.right)

def postorder(n):
    if n is not None:        
        postorder(n.left)
        postorder(n.right)
        print(n.data, end = ' ')

print('전위 순회: ')
preorder(root)
print('\n')
print('중위 순회: ')
inorder(root)
print('\n')
print('후위 순회: ')
postorder(root)
```
